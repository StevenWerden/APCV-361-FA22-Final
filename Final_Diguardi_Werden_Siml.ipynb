## Linear Regression Code

```python
def getData(fileName):
    
    data = pd.read_csv(fileName,skiprows=1, sep='\t', names=['Age', 'Length'])
    df = pd.DataFrame(data, columns= ['Age','Length'])

    sortedDF = df.sort_values(by=['Age'])

    ages = sortedDF['Age'].to_numpy()
    length = sortedDF['Length'].to_numpy()

    return (ages, length)
```

```python
def plotData(xVals, yVals):
    xVals = np.array(xVals)
    yVals = np.array(yVals)
    xVals = xVals
    
    
    plt.plot(xVals, yVals, 'bo',
               label = '')
    plt.show()
```

```python
def labelPlot(x,y,title):
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
```

```python
def chart(df, x, y,title):
    df = df .sort_values([x])
    xVal = df[x]
    yVal = df[y]
    labelPlot(x,y,title)
    plotData(xVal, yVal)
```

```python
def fitData(xVals, yVals):
    xVals = np.array(xVals)
    yVals = np.array(yVals)
    
    plt.plot(xVals, yVals, 'bo',
               label = 'Measured points') 
    
    model = np.polyfit(xVals, yVals, 2)
    
    estYVals = np.polyval(model, xVals)
    
    plt.plot(xVals, estYVals, 'r',
               label = 'Linear fit, k = '
               + str(round(1/model[0], 5)))
    plt.legend(loc = 'best')
    plt.show()
```

```python
def rSquared(observed, predicted):
    error = ((predicted - observed)**2).sum()
    meanError = error/len(observed)
    return 1 - (meanError/np.var(observed))
```

```python
def genFits(xVals, yVals, degrees):
    models = []
    
    if isinstance(degrees, int):
        model = np.polyfit(xVals, yVals, degrees)
        return model
    
    for d in degrees:
        model = np.polyfit(xVals, yVals, d)
        models.append(model)
    return models
```

```python
def testFits(models, degrees, xVals, yVals):
    plt.plot(xVals, yVals, 'o', label = 'Data')
    
    if isinstance(degrees, int):
        estYVals = np.polyval(models, xVals)
        error = rSquared(yVals, estYVals)
        plt.plot(xVals, estYVals,linewidth=3.0,
                   label = 'Fit of degree '\
                   + str(degrees)\
                   + ', R2 = ' + str(round(error, 5)))
        plt.legend(loc = 'best')
        plt.show()
        return
    
    for i in range(len(models)):
        estYVals = np.polyval(models[i], xVals)
        error = rSquared(yVals, estYVals)
        plt.plot(xVals, estYVals,linewidth=3.0,
                   label = 'Fit of degree '\
                   + str(degrees[i])\
                   + ', R2 = ' + str(round(error, 5)))
    plt.legend(loc = 'best')
    plt.show()
```

```python
def linearChart(df, x, y, deg, title):
    df = df .sort_values([x])
    xVal = df[x]
    yVal = df[y]
    degrees = deg
    labelPlot(x,y,title)
    models = genFits(xVal, yVal, degrees)
    testFits(models, degrees, xVal, yVal)
```

```python
def removeOutlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
```

## Cross Validation Code

```python
def splitData(xVals, yVals):
    toTrain = random.sample(range(len(xVals)),
                            len(xVals)//2)
    trainX, trainY, testX, testY = [],[],[],[]
    for i in range(len(xVals)):
        if i in toTrain:
            trainX.append(xVals[i])
            trainY.append(yVals[i])
        else:
            testX.append(xVals[i])
            testY.append(yVals[i])
    return trainX, trainY, testX, testY
```

```python
def runTrials( xVals, yVals, dim, equ, msg):
    #Initialize things 
    numSubsets = 100
    dimensions = dim
    trialList = {}
    
    xVals = np.array(xVals)
    yVals = np.array(yVals)

    for d in dimensions:
        trialList[d] = []
    
    #Train, Test, and Report
    for f in range(numSubsets):
        trainX, trainY, testX, testY = splitData(xVals, yVals)
        for d in dimensions:
            model = np.polyfit(trainX, trainY, d)
            estYVals = np.polyval(model, testX)
            trialList[d].append(equ(testY, estYVals))
            
    print(msg)
    
    for d in dimensions:
        mean = round(sum(trialList[d])/len(trialList[d]), 4)
        sd = round(np.std(trialList[d]), 4)
        print('For dimensionality', d, 'mean =', mean,
              'Std =', sd)
```

```python
def crossValidation(df, x, y, deg, title):
    df = df .sort_values([x])
    xVal = df[x]
    yVal = df[y]
    degrees = deg
    models = genFits(xVal, yVal, degrees)
    runTrials(xVal, yVal, degrees , rSquared, title)  
```

## Bar Charts

```python
def openAll(fileName):
    df = pd.read_csv(fileName)
    return df

def getData(fileName):
    readFile = openAll(fileName)
    return readFile

def getStateYear(readFile, year, state):
    readYear = readFile.loc[lambda readFile: readFile["Year"] == year]
    readStateYear = readYear.loc[lambda readYear: readYear["State"] == state]
    return readStateYear

def sumCells(data, column):
    returnValue = 0
    for cell in data[column]:
        returnValue += cell
    return returnValue

def percentBarGraph2022(readFile):
    stateValues = [[], [], []]
    states = ["California", "Wyoming", "South Carolina"]
    for i in range(len(stateValues)):
        state2022 = getStateYear(readFile, 2022, states[i])
        stateValues[i].append(sumCells(state2022, "Uninsured %"))
        stateValues[i].append(sumCells(state2022, "Unemployed %"))
        stateValues[i].append(sumCells(state2022, "Homeowners %"))
        stateValues[i].append(sumCells(state2022, "Severe Housing Cost Burden %"))
    xvalue = np.arange(4)
    w = 0.2
    ax = plt.subplot(111)
    axisLabels = ["Uninsured", "Unemployed", "Homeowners", "Cost Burden"]
    california = ax.bar(xvalue-0.2, stateValues[0], width = w, align='center', color="blue", edgecolor = "black")
    wyoming = ax.bar(xvalue, stateValues[1], width = w, align='center', color="orange", edgecolor = "black")
    southCarolina = ax.bar(xvalue+0.2, stateValues[2], width = w, align='center', color="green", edgecolor = "black")
    plt.xticks(xvalue, axisLabels)
    ax.legend((california[0], wyoming[0], southCarolina[0]), ('CA', 'WY', 'SC'))
    plt.title("States' % Value of Homemlessness Factors in 2022")
    plt.ylabel("Sum of County Percentages")
    plt.show()

def scatterUninsuredUnemployedNum(readFile):
    states = ["California", "Wyoming", "South Carolina"]
    years = [2019, 2020, 2021, 2022]
    for i in range(len(states)):
        fig, ax = plt.subplots(nrows=2, ncols=2) # For each state, plot the four years in a 2x2 subplot
        for j in range(len(years)): 
            readYear = readFile.loc[lambda readFile: readFile["Year"] == years[j]]
            currentStateX = []
            currentStateY = []
            readStateYear = readYear.loc[lambda readYear: readYear["State"] == states[i]]
            for cell in readStateYear["Uninsured #"]:
                if(cell < 500000): # Filter outlier in CA
                    currentStateX.append(cell)
            for cell in readStateYear["Unemployed #"]:
                if(cell < 200000): # Filter outlier in CA
                    currentStateY.append(cell)
            plt.subplot(2, 2, j + 1)
            plt.scatter(currentStateX, currentStateY)
    
def histHousingBurden2022(readFile):
    states = ["California", "Wyoming", "South Carolina"]
    colors = ["blue", "orange", "green"]
    readYear = readFile.loc[lambda readFile: readFile["Year"] == 2022]
    for i in range(len(states)):
        currentState = []
        readStateYear = readYear.loc[lambda readYear: readYear["State"] == states[i]]
        for cell in readStateYear["Severe Housing Cost Burden %"]:
            currentState.append(cell)
        plt.hist(currentState, color=colors[i], edgecolor = "black")
        plt.title(states[i] + "'s Severe Housing Cost Burden % in 2022")
        plt.xlabel("Housing Cost Burden")
        plt.ylabel("Frequency")
        readStateYear["Severe Housing Cost Burden %"].plot.kde(color='red', secondary_y=True)
        plt.show()
    
def interpretData():
    pd.set_option('display.max_rows', None) # Allows for all rows to be printed
    readFile = getData("Project Data.csv")
    
    # Present the data
    percentBarGraph2022(readFile)
    
    # Analyze the frequency of severe housing cost burden % in 2022 for each state
    histHousingBurden2022(readFile)
```
