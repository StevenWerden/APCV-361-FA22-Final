```python
import matplotlib.ticker as ticker

def chart(data, variables, years, diff = False, state = 'All States'):
    
    tick_spacing = 1
    fig, ax = plt.subplots(1,1)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

    for var in variables:
        stateVars = []
        offset = 0
        msg = var
        
        for year in years:
    
            if state == 'All States':
                temp = data[(data['Year'] == year)]
            else:
                temp = data[(data['Year'] == year) & (data['State'] == state)]
        
            if diff and year == years[0]:
                offset = temp[var].mean()
                msg += ' Diff'
        
            stateVars.append(temp[var].mean() - offset)
        
        plt.plot(years, stateVars, label = msg)
        
    plt.title(state)
    plt.legend(loc = 'best')
    plt.show()
    plt.show()
```

## Linear Regression Code

```python
def getData(fileName):
    
    data = pd.read_csv(fileName,skiprows=1, sep='\t', names=['Age', 'Length'])
    df = pd.DataFrame(data, columns= ['Age','Length'])

    sortedDF = df.sort_values(by=['Age'])

    ages = sortedDF['Age'].to_numpy()
    length = sortedDF['Length'].to_numpy()

    return (ages, length)
```

```python
def plotData(xVals, yVals):
    xVals = np.array(xVals)
    yVals = np.array(yVals)
    xVals = xVals
    
    
    plt.plot(xVals, yVals, 'bo',
               label = '')
    plt.show()
```

```python
def labelPlot(x,y,title):
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
```

```python
def chart(df, x, y,title):
    df = df .sort_values([x])
    xVal = df[x]
    yVal = df[y]
    labelPlot(x,y,title)
    plotData(xVal, yVal)
```

```python
def fitData(xVals, yVals):
    xVals = np.array(xVals)
    yVals = np.array(yVals)
    
    plt.plot(xVals, yVals, 'bo',
               label = 'Measured points') 
    
    model = np.polyfit(xVals, yVals, 2)
    
    estYVals = np.polyval(model, xVals)
    
    plt.plot(xVals, estYVals, 'r',
               label = 'Linear fit, k = '
               + str(round(1/model[0], 5)))
    plt.legend(loc = 'best')
    plt.show()
```

```python
def rSquared(observed, predicted):
    error = ((predicted - observed)**2).sum()
    meanError = error/len(observed)
    return 1 - (meanError/np.var(observed))
```

```python
def genFits(xVals, yVals, degrees):
    models = []
    
    if isinstance(degrees, int):
        model = np.polyfit(xVals, yVals, degrees)
        return model
    
    for d in degrees:
        model = np.polyfit(xVals, yVals, d)
        models.append(model)
    return models
```

```python
def testFits(models, degrees, xVals, yVals):
    plt.plot(xVals, yVals, 'o', label = 'Data')
    
    if isinstance(degrees, int):
        estYVals = np.polyval(models, xVals)
        error = rSquared(yVals, estYVals)
        plt.plot(xVals, estYVals,linewidth=3.0,
                   label = 'Fit of degree '\
                   + str(degrees)\
                   + ', R2 = ' + str(round(error, 5)))
        plt.legend(loc = 'best')
        plt.show()
        return
    
    for i in range(len(models)):
        estYVals = np.polyval(models[i], xVals)
        error = rSquared(yVals, estYVals)
        plt.plot(xVals, estYVals,linewidth=3.0,
                   label = 'Fit of degree '\
                   + str(degrees[i])\
                   + ', R2 = ' + str(round(error, 5)))
    plt.legend(loc = 'best')
    plt.show()
```

```python
def linearChart(df, x, y, deg, title):
    df = df .sort_values([x])
    xVal = df[x]
    yVal = df[y]
    degrees = deg
    labelPlot(x,y,title)
    models = genFits(xVal, yVal, degrees)
    testFits(models, degrees, xVal, yVal)
```
